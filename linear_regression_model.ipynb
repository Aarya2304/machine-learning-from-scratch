```python
import pandas as pd
```

```python
df = pd.read_csv("http://bit.ly/w-data")
print(df.head())
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['Hours'],df['Scores'])
plt.title('Hours Studied vs Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Scores')
plt.grid(True)
plt.show()
```

```python
import numpy as np

X = df['Hours'].values
Y = df['Scores'].values

X = X.reshape(-1,1)
Y = Y.reshape(-1,1)
```

```python
m = 0
b = 0
learning_rate = 0.01
epochs = 1000
n = len(X)
y_pred = m * X + b

loss_history = []

for epoch in range (epochs):
    y_pred = m * X + b
    error = Y - y_pred

    dm = (-2/n) * np.sum(X * error)
    db = (-2/n) * np.sum(error)

    m = m - learning_rate * dm
    b = b - learning_rate * db

    loss  = np.mean(error ** 2)
    loss_history.append(loss)
```

```python
plt.plot(loss_history)
plt.title('Loss History')
plt.xlabel('Epoch')
plt.ylabel('Mean squared errror')
plt.grid(True)
plt.show()
```

```python
plt.scatter(X, Y)

line = m * X + b
plt.plot(X, line)
plt.title('Linear Regression Fit')
plt.xlabel('Hours Studied')
plt.ylabel('Scores')
plt.grid(True)
plt.show()
```

```python
print(f"slope: {m:.4f}")
print(f"Intercept: {b:.4f}")
```

```python
hours = 5
predicted_score = m * hours + b
print(f"Predicted score: {predicted_score:.2f}")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

reg = LinearRegression()
reg.fit(X, Y)

y_sk = reg.predict(X)

print(f"Slope: {reg.coef_[0][0]:.4f}")
print(f"Intercept: {reg.intercept_[0]:.4f}")
print(f"Mean Squared Error: {mean_squared_error(Y,y_sk):.4f}")
```

```python
print(f"Predicted score (sklearn): {reg.predict([[hours]])[0][0]:.2f}")
```
